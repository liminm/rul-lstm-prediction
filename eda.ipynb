```python
# The dataset has 26 columns
# 1. Unit Number (Which engine is it?)
# 2. Time Cycles (How long has it been running?)
# 3-5. Operational Settings (Altitude, Speed, etc.)
# 6-26. Sensor Readings (s1 to s21)

index_names = ['unit_nr', 'time_cycles']
setting_names = ['setting_1', 'setting_2', 'setting_3']
sensor_names = ['s_{}'.format(i) for i in range(1, 22)] 
col_names = index_names + setting_names + sensor_names

print(col_names)
```

```python
import pandas as pd

data_df = pd.read_csv('data/train_FD001.txt', sep=r'\s+', header=None, names=col_names)

print("Training data shape:", data_df.shape)
data_df.head()
```

```python
data_df.describe()
```

```python
drop_cols = ['setting_3', 's_1', 's_5', 's_10', 's_16', 's_18', 's_19']
data_df = data_df.drop(columns=drop_cols)
data_df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Calculate the correlation matrix
corr_matrix = data_df.corr()

# 2. Plot the heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
drop_cols = ['s_14']
data_df = data_df.drop(columns=drop_cols)
data_df.describe()
```

```python
import matplotlib.pyplot as plt

# 1. Isolate data for just one engine (Unit 1)
unit_1_df = data_df[data_df['unit_nr'] == 1]

# 2. Plot Sensor 11 against time (Cycles)
plt.figure(figsize=(10, 6))
plt.plot(unit_1_df['time_cycles'], unit_1_df['s_11'], color='blue')
plt.title('Sensor 11 readings for Unit 1 over its lifetime')
plt.xlabel('Time (Cycles)')
plt.ylabel('Sensor 11 Value')
plt.grid(True)
plt.show()
```

```python
# 1. Calculate the max cycle (total life) for each engine
max_cycles_df = data_df.groupby('unit_nr')['time_cycles'].max().reset_index()
max_cycles_df.columns = ['unit_nr', 'max_life']

# 2. Merge this back into the original dataframe
data_df = data_df.merge(max_cycles_df, on='unit_nr', how='left')
data_df.head()
```

```python
# 3. Calculate RUL
data_df['RUL'] = data_df['max_life'] - data_df['time_cycles']

# 4. Drop the helper column if you want
data_df.drop(columns=['max_life'], inplace=True)

# Verify
data_df[['unit_nr', 'time_cycles', 'RUL']].head()
```

```python
# Clip RUL at 125
data_df['RUL_clipped'] = data_df['RUL'].clip(upper=125)

# Visualize the difference for Unit 1
plt.figure(figsize=(10, 6))
plt.plot(data_df[data_df['unit_nr']==1]['time_cycles'], data_df[data_df['unit_nr']==1]['RUL'], label='True Linear RUL')
plt.plot(data_df[data_df['unit_nr']==1]['time_cycles'], data_df[data_df['unit_nr']==1]['RUL_clipped'], label='Clipped RUL', linewidth=3)
plt.legend()
plt.title('Linear vs Clipped RUL for Unit 1')
plt.show()
```

```python
from torch.utils.data import Dataset, DataLoader


class CMAPSSDataset(Dataset):
    def __init__(self, data_frame: pd.DataFrame, window_size: int =30):
        self.data_frame = data_frame
        self.window_size = window_size
        self.data = []

        for unit_id, group in data_frame.groupby('unit_nr'):
            sensor_data = group.drop(columns=['unit_nr', 'time_cycles', 'RUL', 'RUL_clipped']).values
            rul_data = group['RUL_clipped'].values

            for i in range(len(sensor_data)):
                if i >= window_size:
                    self.data.append((sensor_data[i-window_size:i], rul_data[i-1]))
```
